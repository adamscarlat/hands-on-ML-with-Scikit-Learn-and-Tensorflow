```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_openml
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
```

```python
mnist = fetch_openml('mnist_784', as_frame=True)
```

```python
X = mnist.data
y = mnist.target

sample_size = 5000
y_sampled = y[:sample_size]

tsne = TSNE(n_components=2, init="random", learning_rate="auto", random_state=42)
X2D_sample = tsne.fit_transform(X[:sample_size])
```

```python
# Color for each point is determined by its value from the y array
plt.scatter(X2D_sample[:,0], X2D_sample[:,1], c=y_sampled.astype(np.int8), cmap="jet", alpha=0.5)
plt.axis('off')
plt.colorbar()
```

```python
pca = PCA(n_components=2, random_state=42)
X2D_pca_sampled = pca.fit_transform(X[:sample_size])
```

```python
plt.scatter(X2D_pca_sampled[:,0], X2D_pca_sampled[:,1], c=y_sampled.astype(np.int8), cmap="jet", alpha=0.5)
plt.axis('off')
plt.colorbar()
```
